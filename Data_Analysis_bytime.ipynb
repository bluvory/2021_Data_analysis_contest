```python
import pandas as pd
import numpy as np
import warnings
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
metadata = pd.read_excel('../dataset/1.배달 호출데이터 요약.xlsx', engine='openpyxl', sheet_name='배달호출정보 테이블명세서')
meta = pd.DataFrame(metadata[4:].values, columns=metadata[3:4].values[0])

data = pd.read_csv('../dataset/CATHOLIC_DATA.csv', header=None)
data.columns = meta['영문 컬럼명'].values[:-1]

bucheon = data[(data['DLVR_DSTN_SIGUNGU'] == '부천시') & (data['DLVR_REQUST_STTUS_VALUE'] == 1)]
print(bucheon.shape)

bucheon.head()
```

- DLVR_RCEPT_TIME: 배달접수시간
- DLVR_CARALC_TIME: 배달배차시간
- DLVR_RECPT_TIME: 배달수령시간
- DLVR_COMPT_TIME: 배달완료시간
- DLVR_CANCL_TIME: 배달취소시간

```python
df = pd.DataFrame()

df['date'] = bucheon['PROCESS_DT']
df['food'] = bucheon['DLVR_STORE_INDUTY_NM']
df['dong'] = bucheon['DLVR_DSTN_DONG']

df['receipt'] = bucheon['DLVR_RCEPT_TIME']
df['dispatch'] = bucheon['DLVR_CARALC_TIME']
df['receive'] = bucheon['DLVR_RECPT_TIME']
df['done'] = bucheon['DLVR_COMPT_TIME']
df['cancle'] = bucheon['DLVR_CANCL_TIME']
```

- 배달배차시간과 배달수령시간이 없는 데이터 제외

```python
df = df[(df['receive'] != '\\N') & (df['dispatch'] != '\\N')]
```

```python
def get_time(x):
    h, m, s = x.split()[1].split(':')
    return int(h)*3600 + int(m)*60 + int(s)
```

```python
df['confirm_time'] = abs(df['dispatch'].map(lambda x: get_time(x)) - df['receipt'].map(lambda x: get_time(x)))
df['cook_time'] = abs(df['receive'].map(lambda x: get_time(x)) - df['dispatch'].map(lambda x: get_time(x)))
df['delivery_time'] = abs(df['done'].map(lambda x: get_time(x)) - df['receive'].map(lambda x: get_time(x)))
df['done_time'] = abs(df['done'].map(lambda x: get_time(x)) - df['receipt'].map(lambda x: get_time(x)))
```

- confirm time: 사장님이 접수를 인지하는 시간
- cook time: 요리하는 시간
- delivery time: 배달하는 시간
- done time: 주문 접수부터 배달 완료까지 총 시간

# 중동에서 각 음식마다 배달 소요시간 시간대 그래프

```python
def get_hour(x):
    return pd.Timestamp(x.split(' ')[1]).time().hour

df['hour'] = bucheon['DLVR_RCEPT_TIME'].map(lambda x: get_hour(x))
```

```python
j = df[(df['dong'] == '중동')]

fig = plt.figure(figsize=(20, 15))
sns.barplot(x='food',
            y='delivery_time',
            data=df.groupby(['food'])['delivery_time'].mean().reset_index())
plt.title('중동 음식별 배달하는 시간 평균')
```

```python
j = df[(df['dong'] == '중동')]

fig = plt.figure(figsize=(15, 10))
sns.barplot(x='hour',
            y='delivery_time',
            data=j.groupby(['hour'])['delivery_time'].mean().reset_index())
```

```python
j = df[(df['dong'] == '중동') & (df['food'] == '카페/디저트')]

fig = plt.figure(figsize=(15, 10))
sns.barplot(x='hour',
            y='delivery_time',
            data=j.groupby(['hour'])['delivery_time'].mean().reset_index())
```

```python
j = df[(df['dong'] == '중동')]

fig = plt.figure(figsize=(15, 10))
ax = sns.pointplot(x='hour',
                   y='delivery_time',
                   hue='food',
                   data=j.groupby(['food', 'hour'])['delivery_time'].mean().reset_index())
ax.set_title('중동 음식별 배달 소요 시간')
```

```python
food_list = [('돈까스/일식',0,0),('치킨',0,1),('아시안/양식',0,2),
             ('카페/디저트',1,0),('분식',1,1),('패스트푸드',1,2),
             ('족발/보쌈',2,0),('찜탕',2,1),('한식',2,2),
             ('중식',3,0),('도시락',3,1),('회',3,2),
             ('피자',4,0),('배달전문업체',4,1),('야식',4,2)]
```

```python
fig, axes = plt.subplots(nrows=5, ncols=3, figsize=(20, 40))
for i in food_list:
    data = j[j['food'] == i[0]]
    x = sns.barplot(x = 'hour',
                    y = 'delivery_time',
                    data=data.groupby(['hour'])['delivery_time'].mean().reset_index(), ax=axes[i[1]][i[2]])
    plt.tight_layout()
    axes[i[1]][i[2]].set(title=i[0])
```

- 위의 그래프를 보면 대체적으로 아침과 저녁 때 배달소요시간이 긴 것을 알 수 있음
- 즉 출퇴근 시간에 배달 속도가 느리다는 것을 뜻함
