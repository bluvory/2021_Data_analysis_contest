# 주문량이 가장 많은 시간대 조사

```python
x1 = bucheon['DLVR_DSTN_DONG'].unique()
x1 = x1.tolist()

y1 = bucheon['DLVR_STORE_INDUTY_NM'].unique()
y1 = y1.tolist()
```

## 주문량이 가장 많은 시간대

```python
df1 = pd.DataFrame(columns=y1)

for j in x1 :
    news = bucheon[bucheon['DLVR_DSTN_DONG'] == j]
    mm1 = []
    for i in y1 :
        data = news[news['DLVR_STORE_INDUTY_NM'] == i]
        if len(data) == 0 :
            mm1.append(None)
        else :
            data = data['count'].groupby(data['hour']).sum()
            data = pd.DataFrame({'hour': data.index, 'count': data.values})
            data = data.sort_values(by=['count'],ascending=False)
            mm1.append(int(data.iloc[0,0]))
    df1.loc[j] = mm1
    
df1
```

## 주문량이 두번째로 많은 시간대

```python
df2 = pd.DataFrame(columns=y1)

for j in x1 :
    news = bucheon[bucheon['DLVR_DSTN_DONG'] == j]
    mm1 = []
    for i in y1 :
        data = news[news['DLVR_STORE_INDUTY_NM'] == i]
        if len(data) == 0 :
            mm1.append(None)
        else :
            data = data['count'].groupby(data['hour']).sum()
            data = pd.DataFrame({'hour': data.index, 'count': data.values})
            data = data.sort_values(by=['count'],ascending=False)
            if len(data) <= 1 :
                mm1.append(None)
            else :
                mm1.append(int(data.iloc[1,0]))
    df2.loc[j] = mm1
df2
```

## 전년대비 전체 배송 주문건 (2019.07-11 ~ 2020.07-11)
- 2019년: 07월 ~ 12월
- 2020년: 01월 ~ 11월
- 2020년도는 01월부터 있지만 월을 통일해서 생각함 (07월 ~ 11월)

```python
def get_month(x):
    return int(x.split('-')[1])

df2019['month'] = df2019['PROCESS_DT'].map(lambda x: get_month(x))
df2020['month'] = df2020['PROCESS_DT'].map(lambda x: get_month(x))
```

```python
print('2019년도 전체 배송 주문건:', df2019[df2019['month'] <= 11].shape[0])
print('2020년도 전체 배송 주문건:', df2020[df2020['month'] >= 7].shape[0])
print(round(df2020[df2020['month'] >= 7].shape[0]/df2019[df2019['month'] <= 11].shape[0]*100, 2), '% 증가')
```

## 전년대비 야간 배송 주문건
- 2019년: 07월 ~ 12월
- 2020년: 01월 ~ 11월
- 2020년도는 01월부터 있지만 월을 통일해서 생각함 (07월 ~ 11월)
- 야간배송: 22시 ~ 24시

```python
night2019 = df2019[(df2019['month'] <= 11) & (df2019['hour'] >= 22) & (df2019['month'] <= 24)].shape[0]
night2020 = df2020[(df2020['month'] >= 7) & (df2020['hour'] >= 22) & (df2020['month'] <= 24)].shape[0]

print('2019년도 전체 배송 주문건:', night2019)
print('2020년도 전체 배송 주문건:', night2020)
print(round(night2020/night2019*100, 2), '% 증가')
```

# 요일별 분석

```python
import datetime
```

```python
days = ['월','화','수','목','금','토','일']

def get_week(x):
    d = pd.Timestamp(x)
    return days[datetime.date(d.year, d.month, d.day).weekday()]
```

```python
bucheon['weekday'] = bucheon['DLVR_RCEPT_TIME'].map(lambda x: get_week(x))
bucheon.head()
```

```python
bucheon['weekday'].value_counts()
```

# 배달 시간 분석

```python
metadata = pd.read_excel('../dataset/1.배달 호출데이터 요약.xlsx', engine='openpyxl', sheet_name='배달호출정보 테이블명세서')
meta = pd.DataFrame(metadata[4:].values, columns=metadata[3:4].values[0])

data = pd.read_csv('../dataset/CATHOLIC_DATA.csv', header=None)
data.columns = meta['영문 컬럼명'].values[:-1]

bucheon = data[(data['DLVR_DSTN_SIGUNGU'] == '부천시') & (data['DLVR_REQUST_STTUS_VALUE'] == 1)]
print(bucheon.shape)

bucheon.head()
```

```python
df = pd.DataFrame()

df['date'] = bucheon['PROCESS_DT']
df['food'] = bucheon['DLVR_STORE_INDUTY_NM']
df['dong'] = bucheon['DLVR_DSTN_DONG']

df['receipt'] = bucheon['DLVR_RCEPT_TIME']
df['dispatch'] = bucheon['DLVR_CARALC_TIME']
df['receive'] = bucheon['DLVR_RECPT_TIME']
df['done'] = bucheon['DLVR_COMPT_TIME']
df['cancle'] = bucheon['DLVR_CANCL_TIME']
```

```python
df = df[(df['receive'] != '\\N') & (df['dispatch'] != '\\N')]
```

```python
def get_time(x):
    h, m, s = x.split()[1].split(':')
    return int(h)*3600 + int(m)*60 + int(s)
```

```python
df['confirm_time'] = abs(df['dispatch'].map(lambda x: get_time(x)) - df['receipt'].map(lambda x: get_time(x)))
df['cook_time'] = abs(df['receive'].map(lambda x: get_time(x)) - df['dispatch'].map(lambda x: get_time(x)))
df['delivery_time'] = abs(df['done'].map(lambda x: get_time(x)) - df['receive'].map(lambda x: get_time(x)))
df['done_time'] = abs(df['done'].map(lambda x: get_time(x)) - df['receipt'].map(lambda x: get_time(x)))
```

- confirm time: 사장님이 접수를 인지하는 시간
- cook time: 요리하는 시간
- delivery time: 배달하는 시간
- done time: 주문 접수부터 배달 완료까지 총 시간

- 배달시간이 1분 이하, 2시간(7200초) 이상이면 결측값으로 판단하고 제외한다

```python
df = df[(df['delivery_time']>= 60) & (df['delivery_time'] <= 7200)]
```
